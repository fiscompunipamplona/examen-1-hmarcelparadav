El principio de Huygens establece que cada punto no 
obstruido en un frente de onda actuará como fuente de una
onda esférica secundaria. Teniendo en cuenta el principio de 
Superposición, la amplitud de la onda en cualquier punto más 
allá del frente de onda inicial es la superposición de las 
amplitudes de todas las ondas secundarias.

m, debido a la interacción de un frente de onda plana que

cm (ver figura difraccion.png). Asuma que la amplitud de la
onda es igual 1, la frecuencia (omega) 1 y la longitud de

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, cos

pix = 30
size = 100. 
step = size / pix 
lda = 1 
k = 2.*pi / lda
A = 1.0 
Xi = float(input("Introduzca la coordenada de la onda en X en cm:")) 
Yi = float(input("Introduzca la coordenada de la onda en Y en cm:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    

for yy in range( len(x) ):
    for xx in range( len(y) ):
        z1 = sqrt( (Xi-x[xx])*(Xi-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) )
        z2 = sqrt( ((Xi+0.01)-x[xx])*((Xi+0.01)-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) ) 
        E1=A*cos(z1*k)
        E2=A*cos(z2*k)
        print("Amplitud Onda 1", E1, "Amplitud Onda 2", E2)

```
